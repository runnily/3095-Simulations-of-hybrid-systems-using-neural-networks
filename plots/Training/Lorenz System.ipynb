## Indroduction

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

This is going through the lorenz system model

## Dataframes

This is to see our dataframe

```python
lorenz_df = pd.DataFrame(data = pd.read_csv("../../data/train/lorenz.csv", usecols=[i for i in range (0,7)]))
lorenz_df
```

This is to make a column which group initial_x, initial_y, initial_z together to get a tuple

```python
lorenz_df['initial_x,initial_y,initial_z'] = list(zip(lorenz_df.initial_x, 
                                                      lorenz_df.initial_y,
                                                      lorenz_df.initial_z))
lorenz_df
```

```python
lorenz_df_x1y1 = lorenz_df[lorenz_df['initial_x,initial_y,initial_z'] == (1,1,1)]
lorenz_df_x1y1
```

## Visulising the data

```python
fig, axs = plt.subplots(ncols=2, figsize=(40,10))
dfig = plt.figure()
ax = dfig.add_subplot(111, projection = '3d')
ax.plot(lorenz_df_x1y1['x'], lorenz_df_x1y1['y'], lorenz_df_x1y1['z'], c='skyblue')
sns.lineplot(data=lorenz_df_x1y1, x="time", y="x", ax=axs[0])
sns.lineplot(data=lorenz_df_x1y1, x="time", y="y", ax=axs[1])
```
